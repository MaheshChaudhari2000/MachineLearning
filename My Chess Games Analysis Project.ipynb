 Here's u can tell that this series contains 'white' along with the id Name and we do not want anything extra apart from id name, in that case we can remove extra things using split and strip function.

```python
def parse_values(dataframe_series):
    values=[]
    for i in dataframe_series:
        values.append(i.split(maxsplit=1)[1].strip('""'))
    return values
```

```python
games_df['white_id']=parse_values(games_df['white_id'])
```

```python
games_df['white_id']
```

```python
games_df['black_id']=parse_values(games_df['black_id'])
games_df['event']=parse_values(games_df['event'])
games_df['site']=parse_values(games_df['site'])
games_df['eco']=parse_values(games_df['eco'])
games_df['termination']=parse_values(games_df['termination'])
```

Now let's add another attribute called win/lose telling us  whether i won or lost the game

```python
def add_column(dataframe):
    values=[]
    j=0
    for i in dataframe.black_id:
        if i=='Lovelishkumar':
            if dataframe.at[j,'result']=='Black won':
                values.append('win')
            elif dataframe.at[j,'result']=='White won':
                values.append('lose')
            else:
                values.append('draw')
        else:
            if dataframe.at[j,'result']=='Black won':
                values.append('lose')
            elif dataframe.at[j,'result']=='White won':
                values.append('win')
            else:
                values.append('draw')
        j+=1
            
    return values  
        
        
        
```

```python
games_df['win/lose']=add_column(games_df[['black_id','result']])
```

```python
games_df
```

 Now, we want to add another columns showing My rating before the match, after the match and oponents rating
 

```python
def my_rating(dataframe):
    j=0
    values1=[]
    #values2=[]
    values3=[]
    for i in dataframe.black_id:
        if i == 'Lovelishkumar':
            values1.append(dataframe.black_elo[j])
            #values2.append(dataframe.black_elo[j] + dataframe.black_rating_diff[j])
            values3.append(dataframe.white_elo[j])
        
        else:
            values1.append(dataframe.white_elo[j])
            #values2.append(dataframe.white_elo[j] + dataframe.white_rating_diff[j])
            values3.append(dataframe.black_elo[j])
        j+=1
    return values1,values3
```

```python
games_df['my_rating'],games_df['oponent_rating']=my_rating(games_df)
```

## Asking and Answering Questions
Let's try to answer some questions about our data.

### Q: How many times did i play with Black Pieces and with White Pieces ?

we can use count fuction to answer this question

```python
total_games_as_black=games_df[games_df.black_id=='Lovelishkumar']['black_id'].count()
total_games_as_white=games_df[games_df.white_id=='Lovelishkumar']['white_id'].count()
```

```python
print('I played {} games with black pieces and {} games with white pieces'.format(total_games_as_black,total_games_as_white))
```

### Q: What is my winning rate with White pieces, Black pieces and overall winning rate?

```python
games_as_black=games_df[games_df.black_id=='Lovelishkumar']
winning_rate_as_black=games_as_black[games_as_black.result=='Black won']['black_id'].count()/len(games_as_black)
```

```python
games_as_white=games_df[games_df.white_id=='Lovelishkumar']
winning_rate_as_white=games_as_white[games_as_white.result=='White won']['white_id'].count()/len(games_as_white)
```

```python
overall_winning_rate=games_df[games_df['win/lose']=='win']['win/lose'].count()/len(games_df)
```

```python
print('My winning rate with black pieces is {:.2f}, with white pieces is {:.2f} and overall winning rate is {:.2f}'.format(winning_rate_as_black*100,winning_rate_as_white*100,overall_winning_rate*100))
```

### Q: Does Lovelishkumar has a title?

```python
def yes_no(dataframe):
    j=0
    value=False
    for id in dataframe.black_id:
        if id == 'Lovelishkumar':
            if dataframe.black_title[j]!='no title':
                value=True
                break
        else:
            if dataframe.white_title[j]!='no title':
                value=True 
                break
                
        j+=1
    return value 
    
        
```

```python
if yes_no(games_df):
    print('Lovelishkumar have a title')
else:
    print('Lovelishkumar never had a title')
```

### Q: Does Lovelishkumar ever played with a titled oponent ? if yes how many players ? is/are any of them is a GM?

```python
def yes_no_GM(dataframe):
    j=0
    times=0
    gm=0
    player={}
    player['GM']=None
    player['others']=None
    value=False
    for id in dataframe.black_id:
        if id == 'Lovelishkumar':
            if dataframe.white_title[j]!='no title':
                if dataframe.white_title[j] =='GM':
                    player['GM']=dataframe.white_id[j]
                    times+=1
                    gm+=1
                    value=True
                else:
                    player['other']=dataframe.white_id[j]
                    times+=1
                    value=True
                    
        else:
            if dataframe.black_title[j]!='no title':
                if dataframe.black_title[j] =='GM':
                    player['GM']=dataframe.white_id[j]
                    times+=1
                    gm+=1
                    value=True
                else:
                    player['other']=dataframe.white_id[j]
                    times+=1
                    value=True
                
                
        j+=1
    return value,times,gm,player
    
```

```python
result,times,gm,player=yes_no_GM(games_df)
event=games_df.at[561,'event']

if result:
    if player['GM']!=None:
        print('Yes,Lovelishkumar had played with {}  titled players, {} of them is a GM id_name {} and i played against him during {} event'.format(times,gm,player['GM'],event))
    else:
        print('Yes,Lovelishkumar had played with {}  titled players and none of them is a GM '.format(times))
else:
    print('Lovelishkumar never played with a titled player')
```
